## Mandelbrot set
- for function $f(z) = z^2 + c$, which values of c converge?

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm
```

```python
def create_mandelbrot(min_re, max_re, min_im, max_im, C_iters=1000,max_iter=100,z_max=10):
    """
    min_re - max_re: bounds of real numbers to observe
    min_im - max_im: bounds of imaginary numbers to observe
    C_iters: number of different real and complex values to try
    max_iter: number of iterations before convergence assumed
    z_max: threshold value for convergence
    output: Mandelbrot set, 2-d array of sie (C_iters,C_iters)
    """
    
    re_band = max_re - min_re
    im_band = max_im - min_im

    C_iters = 1000
    max_iter = 100
    z_max = 10

    mandelbrot = np.zeros((C_iters,C_iters))

    for re in range(C_iters):
        for im in range(C_iters):

            c = complex(min_re + (re/C_iters)*re_band,min_im + (im/C_iters)*im_band)
            z = complex(0,0)
            num_iter = 0

            while abs(z) < z_max and num_iter < max_iter:
                z = z**2 + c
                num_iter+=1

            mandelbrot[re, im] = num_iter/max_iter
            
    return mandelbrot
```

```python
min_re, max_re = -1.7,0.5
min_im, max_im = -1.2,1.2

mandelbrot = create_mandelbrot(min_re, max_re,min_im, max_im )

fig = plt.figure(figsize=(10,10))
plt.imshow(mandelbrot, interpolation='nearest',cmap = cm.hot)
```

```python
min_re, max_re = 0.23,0.43
min_im, max_im = -0.2,0.2

mandelbrot = create_mandelbrot(min_re, max_re,min_im, max_im)

fig = plt.figure(figsize=(10,10))
plt.imshow(mandelbrot, interpolation='nearest',cmap = cm.hot)
```
